## QUESTION_05
- Make 5 questions and solve them by using group by.
    - Here are 5 questions I can solve using groupby.
      

### 1:Count number of trials by Status:

```python
df.groupby('Status')['NCT Number'].count()
```

### 2:Get mean enrollment by Age:

```python
df.groupby('Age')['Enrollment'].mean()
```

### 3:Calculate total enrollment by Sponsor:

```python
df.groupby('Sponsor/Collaborators')['Enrollment'].sum()
```

### 4:Get min and max enrollments by Phases:

```python
df.groupby('Phases')['Enrollment'].agg([min, max])
```

### 5:Count trials by Locations:

```python
df.groupby('Locations')['NCT Number'].count()
```

## QUESTION_06
- Make 5 questions and solve them by using conditional statements

## Here are 5 questions that can be solved using conditional statements 
1:Get trials where enrollment is between 500-1000:

```python
df[df['Enrollment'].between(500,1000)]
```

## 2:Get trials in Phase 2 or Phase 3:

```python
df[(df['Phases'] == 'Phase 2') | (df['Phases'] == 'Phase 3')]
```

## 3: Get trials funded by Industry and located in USA:

```python
df[(df['Funded Bys'] == 'Industry') & (df['Locations'].str.contains('USA'))]
```

### 4:Get trials with completion date before 2021:

```python
df[df['Completion Date'] < '2021']
```
